```python
df['station'].unique()


cleaned_df = df[df['station'].map(lambda x: not x.isdigit())]

import json 
def list_to_str(x):
    l = json.loads(x)
    if l == ['']: return []
    return l

cleaned_df['bike_numbers'] = cleaned_df['bike_numbers'].map(list_to_str)
#print(len(df), len(cleaned_df))
```

```python
def something(x):
    x = x.sort_values('time')
    out = []
    arrived_agg = []
    for date, day in x.groupby('date'):
        print(date)
        print('-'*10)

        left, arrived = [], []
        for station_name, station in day.groupby('station'):     

            for(old_idx, old_row), (new_idx, new_row) in zip(station[:-1].iterrows(),station[1:].iterrows()):
                for bike in old_row['bike_numbers']:
                    if not bike in new_row['bike_numbers']:
                        left.append( (bike, new_row['time'], station_name))

                for bike in new_row['bike_numbers']:
                    if not bike in old_row['bike_numbers']:
                        arrived.append( (bike, new_row['time'], station_name))

        left = sorted(left, key=lambda x: x[1])
        arrived = sorted(arrived, key=lambda x: x[1])

        journeys = []
        for (left_bike, left_time, left_station) in left:

            for i, (arrived_bike, arrived_time, arrived_station) in enumerate(arrived):
                 if left_bike == arrived_bike and arrived_time > left_time:
                    journeys.append({
                        'bike_id': left_bike,
                        'left_station': left_station,
                        'arrived_station': arrived_station,
                        'left_time': left_time,
                        'arrived_time': arrived_time
                    })

                    arrived.pop(i)

        out.extend(journeys)
        arrived_agg.append(arrived)
        
    out = pd.DataFrame(out)
        
    return out, arrived_agg

    
```

```python
import time
t = time.time()
raw_journeys, arrived = something(cleaned_df[cleaned_df['time'].map(lambda x: x.month) == 9])
print(time.time() - t)
```
